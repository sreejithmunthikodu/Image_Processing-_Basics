```python
from PIL import Image
import os
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
base_dir = os.path.dirname(os.path.abspath("__file__"))
image_folder = os.path.join(os.path.dirname(base_dir), "Data", "Test")
```

### Import  Images and change format

```python
for file in os.listdir(image_folder):
    if file.endswith(".jpeg"):
        image = Image.open(os.path.join(image_folder, file))
        im_name, ext = os.path.splitext(file)
        image.save(f"{image_folder}/pngs/{im_name}.png")
```

### Resize images keeping aspect ratio
image.thumbnail() # modifies original image

```python
for file in os.listdir(image_folder):
    if file.endswith(".jpeg"):
        image = Image.open(os.path.join(image_folder, file))
        im_name, ext = os.path.splitext(file)
        image.thumbnail((300,300))
        image.save(f"{image_folder}/300_px/{im_name}{ext}")
```

### Resize image discarding aspect ratio
image.resize() # returns new image. Doesnt modify original

```python
for file in os.listdir(image_folder):
    if file.endswith(".jpeg"):
        image = Image.open(os.path.join(image_folder, file))
        im_name, ext = os.path.splitext(file)
        image = image.resize((150,150))
        image.save(f"{image_folder}/300_px/{im_name}new{ext}")
```
